### pip install pykakasi をインストールする

```python
# pykakasi を使って、文字列を英語表記に変更する
from pykakasi import kakasi

kakasi = kakasi()

kakasi.setMode('H', 'a')
kakasi.setMode('K', 'a')
kakasi.setMode('J', 'a')

conv = kakasi.getConverter()
```

```python
import pandas as pd
df = pd.read_excel('sample.xlsx',sheet_name='変数一覧', skiprows=3)
```

```python
# 不要な列を削除(1～3列目)
df = df.drop(columns=df.columns[[0, 1, 2]])
df
```

```python
before = df['氏名']
before
```

```python
# 列名 Before, Afterで空のDataframeを作成する
list_df = pd.DataFrame( columns=['Before','After'] )
```

```python
# 変数beforeを、pykakasi を使って、日本語からローマ字に変換して、Dataframeにインサートする
for i, row in enumerate(before, 1):
    tmp_se = pd.Series( [ row, conv.do(row) ], index=list_df.columns , name='No.' + str(i) )
    list_df = list_df.append( tmp_se)

# 変換したDataframeの結果を出力する
list_df
```

```python
# Excelファイルを出力
list_df.to_excel('output.xlsx')
```

# 複数のシートを読み込む場合

```python
# 複数シートを読み込む
import pandas as pd

df = pd.read_excel('code.xlsx', sheet_name = None , header = None, skiprows=3)

# シートデータを、sheetvalues_listに設定する(※複数データが登録される)
sheetvalues_list = list(df.values())
# sheet4番目のデータを表示する
sheetvalues_list[3]
```

```python
# sheet名を表示する
sheetname_list = list(df.keys())
print("・シート名のリスト",sheetname_list)
```

```python
# 空のdf_allのDataFrameを作成する
df_all = pd.DataFrame()
```

```python
# sheetname_listのシート名をもとに、for文を回して、必要なデータだけをdf_allに設定する
for i, sheetname in enumerate(sheetname_list):
    # 見出し、修正履歴を処理させる
    if sheetname not in ("見出し","修正履歴"):
        # df_allに見出し、修正履歴以外のデータを登録する
        df_all = pd.concat([df_all, sheetvalues_list[i]])
```

```python
df_all.head(10)
```

```python
# 必要な項目だけを抽出する
df_all[[4,5,6]].head(7)
```

```python
# １．売上日列に、売り上げという文字が入っていない行を抽出
# ２．必要な項目だけを抽出する。
df_dummy = df_all[[5,6]][(df_all[4]!="売上日")]
```

```python
df_dummy.head()
```

```python
before = df_dummy[6]
before
```

```python
# 列名 Before, Afterで空のDataframeを作成する
list_df = pd.DataFrame( columns=['Before','After'] )
```

```python
# pykakasi を使って、文字列を英語表記に変更する
from pykakasi import kakasi

kakasi = kakasi()

kakasi.setMode('H', 'a')
kakasi.setMode('K', 'a')
kakasi.setMode('J', 'a')

conv = kakasi.getConverter()
```

```python
# 変数beforeを、pykakasi を使って、日本語からローマ字に変換して、Dataframeにインサートする
for i, row in enumerate(before, 1):
    tmp_se = pd.Series( [ row, conv.do(row) ], index=list_df.columns , name='No.' + str(i) )
    list_df = list_df.append( tmp_se)

# 変換したDataframeの結果を出力する
list_df
```

```python
# Excelファイルを出力
list_df.to_excel('output.xlsx')
```
