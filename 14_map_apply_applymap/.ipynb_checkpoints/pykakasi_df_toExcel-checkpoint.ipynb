```python
# pykakasi を使って、文字列を英語表記に変更する
from pykakasi import kakasi

kakasi = kakasi()

kakasi.setMode('H', 'a')
kakasi.setMode('K', 'a')
kakasi.setMode('J', 'a')

conv = kakasi.getConverter()
```

```python
import pandas as pd
```

```python
df = pd.read_excel('sample.xlsx', skiprows=3)
df.head()
```

```python
df = df.drop(columns=df.columns[[0,1,2]])
df
```

```python
# Seriesの各要素に対してはmap関数を
f = lambda x: conv.do(x)

df['ローマ字'] = df['氏名'].map(f)
df
```

```python
df.to_excel('df_output.xlsx')
```

```python
df
```

```python
df['結果'] = df[['ローマ字']][(df['社員ID'] == 'a023') & (df['商品分類'] == 'ボトムス')]
```

```python
# 欠損値だけの行を抽出
df[df['結果'].isnull() ].head()
```

```python
# 欠損値以外の行を抽出
df[~df['結果'].isnull()].head()
```
