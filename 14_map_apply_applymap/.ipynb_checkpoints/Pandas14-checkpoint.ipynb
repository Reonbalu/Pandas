```python
import pandas as pd
import numpy as np
```

# Seriesとmapメソッド

```python
s1 = pd.Series([7000, 5000, 23000, 2500, 12000])
s1
```

```python
s1 * 2
```

```python
# 2倍にする関数
def double(x):
    return x * 2
```

```python
s1.map(double)
```

```python
s1.map(lambda x: x * 2)
```

```python
s1.map(lambda x: '1万円以上' if x >= 10000 else '1万円未満')
```

```python
# 価格でランク付をする関数
def f_rank(x):
    if x >= 20000:
        return 'S'
    elif x >= 10000:
        return 'A'
    elif x >= 5000:
        return 'B'
    else:
        return 'C'
```

```python
s1.map(f_rank)
```

```python
s1.map(np.square)
```

```python
s1.map(np.sum)
```

```python
s1.sum()
```

### 文字列Seriesとmapメソッド

```python
s2 = pd.Series(['スカートSkirt', 'ニットKnit', 'ジャケットJacket', 'シャツShirt', 'ロングパンツSlacks'])
s2
```

```python
s2.map(lambda x: x[0])
```

```python
import re
s2.map(lambda x: re.findall('[A-z]+', x))
```

```python
s3 = pd.Series(['スカートSkirt', 'ニットKnit', 'ジャケットJacket', 'シャツShirt', 'BottomsロングパンツSlacks'])
s3
```

```python
s3.map(lambda x: re.findall('[A-z]+', x))
```

```python
s4 = pd.Series(['スカートSkirt', 'ニットKnit', 'ジャケットJacket', 'シャツShirt', np.nan])
s4
```

```python
d = {'ジャケットJacket': 'Outer', 'スカートSkirt': 'Bottoms', 'ニットKnit': 'Tops','ワンピースOnepiece':'Onepiece' }
d
```

```python
s4.map(d)
```

```python
s4.map('{}を買います。'.format)
```

```python
s4.map('{}を買います。'.format, na_action='ignore')
```

# Seriesとapplyメソッド

```python
s1
```

```python
s1.apply(double)
```

```python
s1.apply(lambda x: x*2)
```

```python
s1.apply(lambda x: '1万円以上' if x >= 10000 else '1万円未満')
```

```python
s1.apply(f_rank)
```

```python
# 消費税率で税込金額を計算
def f_tax(x, tax):
    return x * tax + x
```

```python
s1.apply(f_tax, args=(0.1,))
```

```python
s1.map(f_tax, args=(0.1,))
```

```python
s1.apply(np.square)
```

```python
s1.apply(np.sum)
```

# DataFrameとapplymap

```python
df = pd.DataFrame(
    [[11000, 6000, 8000],[5000, 12000, 6000],[4000, 5000, 9000]],
    columns=['1日', '2日', '3日'], index=['Aさん', 'Bさん', 'Cさん'] )
df
```

```python
df * 2
```

```python
df.applymap(double)
```

```python
df.applymap(lambda x : x * 2)
```

```python
df.applymap(lambda x: '1万円以上' if x >= 10000 else '1万円未満')
```

```python
df.applymap(f_rank)
```

```python
df.applymap(f_tax, args=(0.1,))
```

### 参考まで

```python
df.mask(df < 10000, '1万円未満')
```

```python
df < 10000
```

```python
df.where(df < 10000, '1万円以上')
```

```python
df.applymap(np.square)
```

```python
df.applymap(np.sum)
```

```python
df.sum()
```
